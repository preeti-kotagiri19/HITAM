### Array:

```
input : [1,2,3,2,1,4,5,5,4]
output : 3
```

### Algorithm:
1. Read multiple integer values from the user at run time
2. val = 0
3. Traverse through array elements(i) and apply ^ on val and i
4. update val value with val ^ i
5. print(val) after all the iterations 

````python
li = list(map(int,input().split()))
val = 0
for i in li:
    val = val ^ i
print(val)
````

### Plus One 

````python
li = list(map(int,input().split()))
# part-1 : convert integer array to a number
num = 0
for ele in li:
    num = (num * 10) + ele
# add 1 to a number
num += 1
# part-2 : convert number to an array
res = []
while num > 0:
    res.insert(0,num % 10)
    num = num // 10
print(res)
````

### Diagonal Difference

````python
n = int(input())
arr = []
for _ in range(n):
    arr.append(list(map(int,input().split())))
d1,d2 = 0,0
for i in range(n):
    for j in range(n):
        if i == j:
            d1 += arr[i][j]
        if j == (n-i-1):
            d2 += arr[i][j]
print(abs(d1-d2))
````

````python
n = int(input())
arr = []
for _ in range(n):
    arr.append(list(map(int,input().split())))
d1,d2 = 0,0
for i in range(n):
    d1 += arr[i][i]
    d2 += arr[i][n-i-1]
print(abs(d1-d2))
````

### Reverse a string without using built-in functions and [::-1]

````python
'''
s = "abc"
output: "cba"

s1 = "python"
output : "nohtyp"
'''
s = input()
li = list(s)
left,right = 0,len(s)-1
while left < right:
    li[left],li[right] = li[right],li[left]
    left += 1
    right -= 1
print("".join(li))
````

### Reverse Vowels of a String 


````python
s = input()
li = list(s)
vowels = "aeiouAEIOU"
left,right = 0,len(s)-1
while left < right:
    while left < right and li[left] not in vowels:
        left += 1
    while left < right and li[right] not in vowels:
        right -= 1
    li[left],li[right] = li[right],li[left]
    left += 1
    right -= 1
print("".join(li))
````

### Search Insert Position 

````python
nums = list(map(int,input().split()))
target = int(input())
left,right = 0,len(nums)-1
while left <= right:
    mid = (left + right) // 2
    if nums[mid] == target:
        print(mid)
        break
    elif target > nums[mid]:
        left = mid + 1
    else:
        right = mid - 1
else:
    print(left)
````
