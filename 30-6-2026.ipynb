### Question:
```
input : [1,2,3,2,1,4,5,4,5]
output : 3
``` 

````python
li = list(map(int,input().split()))
d = {}
for ele in li:
    if ele not in d:
        d[ele] = 1
    else:
        d[ele] += 1
for i in d:
    if d[i] == 1:
        print(i)
        break
````

````python
li = list(map(int,input().split()))
d = {}
for ele in li:
    d[ele] = d.get(ele,0) + 1
for i in d:
    if d[i] == 1:
        print(i)
        break
````

````python
li = list(map(int,input().split()))
val = 0
for ele in li:
    val = val ^ ele
print(val)
````

### Plus One 


````python
digits = list(map(int,input().split()))
#part-1 : convert a list to a number
num = 0
for d in digits:
    num = (num * 10) + d
#increment num by 1
num += 1
#part-2 : convert a number to a list
res = []
while num > 0:
    res.insert(0,(num % 10))
    num = num // 10
print(res)
````

### Diagonal Difference

````python
n = int(input())
arr = []
for _ in range(n):
    arr.append(list(map(int,input().split())))
d1,d2 = 0,0
for row in range(n):
    for col in range(n):
        if row == col:
            d1 += arr[row][col]
        if (row + col) == (n-1):
            d2 += arr[row][col]
print(abs(d1 - d2))
````

````python
n = int(input())
arr = []
for _ in range(n):
    arr.append(list(map(int,input().split())))
d1,d2 = 0,0
for row in range(n):
    d1 += arr[row][row]
    d2 += arr[row][n-1-row]
print(abs(d1 - d2))
````

### Question: Rotate Array

````python
nums = [1,2,3,4,5,6,7]
k = 3
for _ in range(k):
    val = nums.pop()
    nums.insert(0,val)
print(nums)
````

````python
nums = [1,2,3,4,5,6,7]
k = 3
n = len(nums)
res = nums[n-k : ]
del nums[n-k : ]
nums = res + nums
print(nums)
````
