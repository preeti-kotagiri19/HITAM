### plus one

### Solution-1:

```python
digits = list(map(int,input().split()))
#convert list to a number
num = 0
for d in digits:
    num = (num * 10) + d
num += 1
#convert number to a list
res = []
while num > 0:
    res.insert(0,num % 10)
    num //= 10
print(res)
```

### Solution-2:

```python
digits = list(map(int,input().split()))
n = len(digits)
for i in range(n-1,-1,-1):
    digits[i] += 1
    digits[i] %= 10
    if digits[i] != 0 :
        print(digits)
        break
else:
    print([1] + digits)
```

### Two sum

```python
nums = [2,7,11,15]
target = 26
n = len(nums)
for i in range(n-1):
    for j in range(i+1,n):
        if nums[i] + nums[j] == target:
            print([i,j])
            break
```

```python
nums = [2,7,11,15]
target = 26
d = {}
for i in range(len(nums)):
    comp = target - nums[i]
    if comp not in d:
        d[nums[i]] = i
    else:
        print([d[comp],i])
        break
        
```

### Diagonal difference:

```python
n = int(input())
arr = []
for _ in range(n):
    arr.append(list(map(int,input().split())))
d1,d2 = 0,0
for row in range(n):
    for col in range(n):
        if row == col:
            d1 += arr[row][col]
        if row + col == (n-1):
            d2 += arr[row][col]
print(abs(d1 - d2))
```

```python
n = int(input())
arr = []
for _ in range(n):
    arr.append(list(map(int,input().split())))
d1,d2 = 0,0
for row in range(n):
    d1 += arr[row][row]
    d2 += arr[row][n-1-row]
print(abs(d1 - d2))
```

### Matrix Diagonal Sum

```python
n = int(input())
arr = []
for _ in range(n):
    arr.append(list(map(int,input().split())))
s = 0
for row in range(n):
    s += arr[row][row]
    s += arr[row][n-1-row]

if n % 2 == 1:
    s -= arr[n//2][n//2]
print(s)
```

### Move hash to front

```python
st = input()
res = ""
for ch in st:
    if ch == "#":
        res = "#" + res
    else:
        res = res + ch
print(res)
```

### Reduce string with repeated characters:

```python
st = input()
res = ""
count = 1
for i in range(1,len(st)):
    if st[i] == st[i-1]:
        count += 1
    else:
        if count > 1:
            res = res + st[i-1] + str(count)
        else:
            res += st[i-1]
        count = 1
res = (res + st[i] + str(count)) if count > 1 else (res + st[i])
print(res)
```

### Reverse each word in a sentence:

```python
st = input()
res = [word[::-1] for word in st.split()]
print(" ".join(res))
```

### Reverse a string without using reversed() and [::-1]

### Search Insert Position 


```python
nums = [1,3,5,6]
target = 7
left,right = 0,len(nums)-1
while left <= right:
    mid = (left + right) // 2
    if target == nums[mid]:
        print(mid)
        break
    elif target > nums[mid]:
        left = mid + 1
    else:
        right = mid - 1
else:
    print(left)
```
