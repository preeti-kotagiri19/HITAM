### Problem-1
```
input : [1,2,1,3,2,4,5,4,5]
output : 3
```
### Algorithm:
1. Read n integer values from the user at run time
2. val = 0
3. Traverse array elements(i) and perform val ^ i and update val value
4. print val after all the iterations

````python
li = list(map(int,input().split()))
val = 0
for i in li:
    val = val ^ i
print(val)
````

### Plus One 
#### Traditional approach:

````python
# part-1: converting list to a number
digits = list(map(int,input().split()))
num = 0
for i in digits:
    num = (num * 10) + i
# add 1 to num
num += 1
# part-2 : convert a number to list
res = []
while num > 0:
    res.insert(0,num % 10)
    num = num // 10
print(res)
````

### pythonic-way:

````python
digits = list(map(int,input().split()))#[1,2,3]
# part-1: converting list to a number
num = int("".join(list(map(str,digits))))
# add 1 to num
num += 1
# part-2 : convert a number to list
res = [int(i) for i in str(num)]
print(res)
````

### Diagonal Difference


````python
n = int(input().strip())
arr = []
for _ in range(n):
    arr.append(list(map(int, input().rstrip().split())))
d1 = 0
d2 = 0
for i in range(n):
    d1 += arr[i][i]
    d2 += arr[i][n-i-1]
print(abs(d1-d2))
````

### Reverse Vowels of a String

````python
st = input()
left,right = 0,len(st)-1
li = list(st)
vowels = "aeiouAEIOU"
while left < right:
    while left < right and li[left] not in vowels:
        left += 1
    while left < right and li[right] not in vowels:
        right -= 1
    li[left],li[right] =li[right],li[left]
    left += 1
    right -= 1
print("".join(li))
````

### Breaking the Records

````python
n = int(input().strip())
scores = list(map(int, input().rstrip().split()))

def breakingRecords(scores):
    max_val,min_val = scores[0],scores[0]
    c1,c2 = 0,0
    for i in range(1,n):
        if scores[i] > max_val:
            max_val = scores[i]
            c1 += 1
        elif scores[i] < min_val:
            min_val = scores[i]
            c2 += 1
    return c1,c2
breakingRecords(scores)
````

### Longest Common Prefix

````python
n = int(input())
li = []
for _ in range(n):
    li.append(input())
li.sort()
first = li[0]
last = li[-1]
m = min(len(first),len(last))
res = ""
for i in range(m):
    if first[i] == last[i]:
        res += first[i]
    else:
        break
print(res)
    
````

### Recursive Digit Sum

````python
n = input()
k = int(input())
def super_digit(n,k):
    digit_sum = sum([int(i) for i in n])
    total_sum = digit_sum * k
    def recursion(num):
        if num < 10:
            return num
        return recursion(sum([int(i) for i in str(num)]))
    return recursion(total_sum)
print(super_digit(n,k))
````
