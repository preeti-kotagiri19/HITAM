### Questions:

```
input : [1,2,3,2,1,4,5,5,4]
output : 3
```

````python
li = list(map(int,input().split()))
d = {}
for ele in li:
    if ele not in d:
        d[ele] = 1
    else:
        d[ele] += 1
for k in d:
    if d[k] == 1:
        print(k)
        break
````

### Plus One 

````python
digits = list(map(int,input().split()))
#part-1 : convert list to a number
num = 0
for d in digits:
    num = (num * 10) + d
num += 1
#part-2 : convert number to a list
res = []
while num > 0:
    res.insert(0,num % 10)
    num //= 10
print(res)
    
````

````python
digits = list(map(int,input().split()))
def plusone(digits):
    for i in range(len(digits)-1,-1,-1):
        digits[i] += 1
        digits[i] %= 10
        if digits[i] % 10 != 0:
            return digits
    return [1] + digits
plusone(digits)
````

### Diagonal Difference

````python
n = int(input())
arr = []
for _ in range(n):
    arr.append(list(map(int,input().split())))
d1,d2 = 0,0
for row in range(n):
    for col in range(n):
        if row == col:
            d1 += arr[row][col]
        if (row + col) == (n-1):
            d2 += arr[row][col]
print(abs(d1 - d2))
    
````

````python
n = int(input())
arr = []
for _ in range(n):
    arr.append(list(map(int,input().split())))
d1,d2 = 0,0
for i in range(n):
    d1 += arr[i][i]
    d2 += arr[i][n-1-i]
print(abs(d1 - d2))
````

### Reverse a string
without using reversed() and [::-1]

```
input :  'python'
output : 'nohtyp'
```

````python
st = input()
li = list(st)
left,right = 0,len(st)-1
while left < right:
    li[left],li[right] = li[right],li[left]
    left += 1
    right -= 1
print("".join(li))
````

````python
s = input()
li = list(s)
left,right = 0,len(s)-1
vowels = "aeiouAEIOU"
while left < right:
    while left < right and li[left] not in vowels:
        left += 1
    while left < right and li[right] not in vowels:
        right -= 1
    li[left],li[right] = li[right],li[left]
    left += 1
    right -= 1
print("".join(li))
    
````
