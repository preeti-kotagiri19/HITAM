```
Question:

input : [1,2,1,3,4,5,4,5,2]
output : 3

Algorithm:

step-1: Read multiple values from the user at run time
step-2 : val = 0
step-3 : Traverse through a list using 'i' pointer
step-4 : apply ^ operator on val and i and update the result in val
step-5 : after all the iterations diplay val value
```

map(function_name,iterable_object)

````python
arr = list(map(int,input().split()))
val = 0
for i in arr:
    val = val ^ i
print(val)
````

```
input : 'abc123def456ghi"
output: 21

step-1 : Read a string from user at run time
step-2 : s = 0
step-3 : Traverse characters in a string
step-4 : check whether the character is alphabet or digit
step-5 : if character is digit,add that digit to your s variable and update s
step-6 : print s value at the end
```

````python
st = input()
s = 0
for ch in st:
    if ord(ch) >= 48 and ord(ch) <= 57:
        s += int(ch)
print(s)
````

````python
st = input()
s = 0
for ch in st:
    if ch.isdigit():
        s += int(ch)
print(s) 
````

````python
def superDigit(n):
    if int(n) >= 0 and int(n) <= 9:
        return int(n)
    s = sum(list(map(int,list(n))))
    return superDigit(str(s))
````

````python
n = "148"
k = 3
st = n * k
superDigit(st)
````

````python
s = "abcdefghaabca"
d = {}
for ch in s:
    if ch not in d:
        d[ch] = 1
    else:
        d[ch] += 1
print(d)
````

````python
s = "abcdefghaabca"
d = {}
for ch in s:
    d[ch] = d.get(ch,0) + 1
print(d)
````

````python
repeated = [(ch, cnt) for ch, cnt in d.items() if cnt > 1]

repeated.sort(key=lambda x: (-x[1],x[0]))

result = ''.join(ch for ch, cnt in repeated)

print(result)
````
