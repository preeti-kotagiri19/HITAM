### Plus One 
```
digits = [1,2,3]
Output: [1,2,4]

digits = [9]
Output: [1,0]
```

````python
digits = list(map(int,input().split()))
#part-1 : convert list to a number
num = 0
for d in digits:
    num = (num * 10) + d
# add 1 to number and update
num += 1
#part-2 : convert a number to list
res = []
while num > 0:
    res.insert(0,num % 10)
    num //= 10
print(res)
````

### Diagonal Difference


````python
n = int(input())
arr = []
for _ in range(n):
    arr.append(list(map(int,input().split())))
d1,d2 = 0,0
for i in range(n):
    d1 += arr[i][i]
    d2 += arr[i][n-i-1]
print(abs(d1 - d2))
````

### Question:
Reverse a string without using reversed() and [::-1]
```
input : "python"
output : "nohtyp"
```

````python
st = input()
"".join(reversed(st))
````

````python
st = input()
left,right = 0,len(st)-1
li = list(st)
while left < right:
    li[left],li[right] = li[right],li[left]
    left += 1
    right -= 1
print("".join(li))
          
````

### Reverse Vowels of a String 

````python
st = input()
li = list(st)
left,right = 0,len(st)-1
vowels = 'aeiouAEIOU'
while left < right:
    while left < right and li[left] not in vowels:
        left += 1
    while left < right and li[right] not in vowels:
        right -= 1
    li[left],li[right] = li[right],li[left]
    left += 1
    right -= 1
print("".join(li))
````

### Question:
```
input :  orange
output : puboif

input :  mahesh
output : obifui

input :  bhavani
output : eibaboj
```

````python
st = input()
res = ""
for ch in st:
    if ch in "aeiou":
        res += chr(ord(ch) + 1)
    else:
        if 'a' < ch < 'e':
            res += 'e'
        elif 'e' < ch < 'i':
            res += 'i'
        elif 'i' < ch < 'o':
            res += 'o'
        elif 'o' < ch < 'u':
            res += 'u'
        elif 'u' < ch <= 'z':
            res += 'a'
print(res)
````

### Search Insert Position 

````python
nums = [1,3,5,6]
target = 2
left,right = 0,len(nums)-1
while left <= right:
    mid = (left + right) // 2
    if target == nums[mid]:
        print(mid)
        break
    elif target > nums[mid]:
        left = mid + 1
    else:
        right = mid - 1
else:
    print(left)
````
