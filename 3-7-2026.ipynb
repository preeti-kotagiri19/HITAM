### Reverse Vowels of a String

````python
st = input()
li = list(st)
left,right = 0,len(st)-1
vowels = "aeiouAEIOU"
while left < right:
    while left < right and li[left] not in vowels:
        left += 1
    while left < right and li[right] not in vowels:
        right -= 1
    li[left],li[right] = li[right],li[left]
    left += 1
    right -= 1
print("".join(li))
    
    
````

````python
st = input()
res = ""
vowels = "aeiou"
for ch in st:
    if ch in vowels:
        res += chr(ord(ch) + 1)
    else:
        if 'a' < ch < 'e':
            res += 'e'
        elif 'e' < ch < 'i':
            res += 'i'
        elif 'i' < ch < 'o':
            res += 'o'
        elif 'o' < ch < 'u':
            res += 'u'
        elif 'u' < ch <= 'z':
            res += 'a'
print(res)
````

###  Longest Common Prefix

````python
#Input: strs = ["flower","flow","flight"]
#Output: "fl"
strs = ["flower","flow","flight"]
strs.sort()
first = strs[0]
last = strs[-1]
res = ""
n = min(len(first),len(last))
for i in range(n):
    if first[i] == last[i]:
        res += first[i]
    else:
        break
print(res)
````

````python
li1 = [1,2,3,4,5]
li2 = [10,20,30,40]
#output : [11,22,33]
res = []
for i in range(len(li1)):
    res.append(li1[i] + li2[i])
print(res)
````

````python
li1 = [1,2,3,4,5]
li2 = [10,20,30,40]
#output : [11,22,33]
res = []
for x,y in zip(li1,li2):
    res.append(x + y)
print(res)
````

### Sorting:
1. Bubble
2. selection
3. Insertion
4. Merge
5. Quick

````python
nums1 = [1,2,3,0,0,0]
m = 3
nums2 = [2,5,6]
n = 3
i,j,k = m-1,n-1,m+n-1
while i >= 0 and j >= 0:
    if nums1[i] > nums2[j]:
        nums1[k] = nums1[i]
        i -= 1
        k -= 1
    else:
        nums1[k] = nums2[j]
        j -= 1
        k -= 1
print(nums1)
````
